```python
# import libraries
import pandas as pd
import numpy as np 
import tensorflow as tf 
import tensorflow_decision_forests as tfdf

```

```python
#load data 
data = pd.read_csv("Data/card_transdata.csv")
```

```python
data
```

We have 1000000 records of data for use.

```python
#renaming columns names for simplicity
data.rename(columns = {'distance_from_home':'dist_f_home','distance_from_last_transaction':'dist_f_last_trans'})
```

 Will use 80% for training,20% for testing. Will take 20 records out of testing for demonstation.

```python
#shuffles the data
data_shuffled = data.sample(frac=1,random_state= 50)


#percent of data used as test split
split_percent = 0.8



#get the index to split the data
total_rows = data.shape[0]
train_size = int(total_rows*split_percent)

#train and test data
train = data_shuffled.iloc[:train_size]
test = data_shuffled.iloc[train_size:]
```

```python
app_test = test.iloc[:20]
test.drop(test.index[:20],inplace=True)
```

Data has been checked and "test" is the data that will be used for tesing, "train" for training, and "app_test" is the 20 records for the demonstration.

```python
label ='fraud'
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train,label=label)
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test,label=label)
```
